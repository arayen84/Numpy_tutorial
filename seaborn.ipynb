```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import random
import numpy as np
```

```python
sns.distplot([1,2,4,6,9],hist=True)
plt.show()
```

# Normal Distribution (Gaussian)


loc - find mean value

scale - standard deviation

size - size of array

```python
sns.distplot(random.normal(loc=2,scale=4,size=(2,3)),hist=True)
plt.show()
```

# Binomial DIstribution



n - no of trial

p - probability

size - size of array

```python
x1=random.binomial(n=31,p=0.5,size=8)
sns.distplot(x1)
plt.show
print(x1)
```

# Multinomial Distribution

n - no of possible outcomes

pvals - list of probability

size - size of array

```python
x2=random.multinomial(n=6,pvals=[1/6,1/6,1/6,1/6,1/6,1/6],size=2)
print(x2)
```

# Exponential distribution

scale - inverse rate = 0.1

```python
x3 = random.exponential(scale=3,size=4)
print(x3)
```

# Uniform distribution

a - lower bound - 0.0

b - upper bound - 1.0

size - size of array

```python
x4=random.uniform(size=(2,3))
sns.distplot(x4)
plt.show()
print(x4)
```

# Logistic Distribution

loc - mean

scale - standard deviation

size - size of array


```python
x5 = random.logistic(loc=3,scale=2,size=(3,4))
sns.distplot(x5,hist=True)
plt.show()
print(x5)
```

# Poission Distribution

lam - rate of occurance

size - size of array

Ex: if i'm doing workout 2 times a day regulary if i do 3 times what will happen?

For that we are using poisson distribution

```python
x6 = random.poisson(lam =2 ,size=5)
sns.distplot(x6,hist=True)
plt.show()
print(x6)
```

# Universal function (ufuncs)



vectorization - cpu faster

```python
y = [1,2,3,4,5]
y1 = [6,7,8,9,1]
z = np.add(y,y1)
print(z)
```

Rounding Decimal

trunc()  and fix() and around()


```python
y2=np.trunc([2.345,-1.874])
print(" Using trunc function :",y2)
print("_______________")
y3 = np.fix([-4.987,0.879])
print(" Using fix function :",y3)
print("_______________")
y4 = np.around([4.498,5.89,-8.1])
print(" Using around function :",y4)
```
